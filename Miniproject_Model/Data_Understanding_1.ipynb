```python
n_cols = 3
n_rows = (len(object_d)+n_cols-1)// n_cols
fig , axes=plt.subplots(n_rows,n_cols, figsize = (5*n_cols,5*n_rows))
for index,column in enumerate(object_d):
    row = index//n_cols
    col = index%n_cols
    sns.countplot(data=data,x=column,ax=axes[row,col])
    axes[row,col].set_title(f'{column}')
    axes[row,col].set_xlabel('')
    axes[row,col].set_ylabel('count')
if len(object_d) < n_rows * n_cols:
    for idx in range(len(object_d), n_rows * n_cols):
        fig.delaxes(axes.flatten()[idx])

plt.tight_layout()
plt.show()
```

```python
df_copy = data.copy()
bins = [18, 25, 40, 60, 90]
labels = ['18-25', '26-40', '41-60', '60+']
df_copy['Age Group'] = pd.cut(df_copy['Age'], bins=bins, labels=labels, right=False)
color_t  = 'flare'
# Create the countplot with custom colors
plt.figure(figsize=(7, 5))
sns.countplot(data=df_copy, x='Age Group', hue='Heart_Attack_Risk', palette=color_t)
plt.title('Heart_Attack_Risk by Age Group')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend(title='Heart_Attack_Risk')
plt.show()
```

```python
bins = [ 149, 200,239, 301]
labels = ['<200', '200-239', '240+']
df_copy['Cholesterol_Level Group'] = pd.cut(df_copy['Cholesterol_Level'], bins=bins, labels=labels, right=False)

# Create the countplot with custom colors
plt.figure(figsize=(7, 5))
sns.countplot(data=df_copy, x='Cholesterol_Level Group', hue='Heart_Attack_Risk', palette=color_t)
plt.title('Heart_Attack_Risk by Cholesterol_Level Group')
plt.xlabel('Cholesterol_Level')
plt.ylabel('Count')
plt.legend(title='Heart_Attack_Risk')
plt.show()
```

```python
df_copy['Smoking Group'] = df_copy['Smoking'].map({0: 'not Smoking', 1: 'Smoking'})
# นับจำนวนทั้งหมดในแต่ละกลุ่ม (สูบ vs ไม่สูบ)
total_counts = df_copy['Smoking Group'].value_counts()


# Create the countplot with custom colors
plt.figure(figsize=(7, 5))
sns.countplot(data=df_copy, x='Smoking Group', hue='Heart_Attack_Risk', palette=color_t)
plt.title('Heart_Attack_Risk by Smoking')
plt.xlabel('Smoking')
plt.ylabel('Count')
plt.legend(title='Heart_Attack_Risk')
plt.show()
```

```python
df_copy['Smoking Group'] = df_copy['Smoking'].map({0: 'not Smoking', 1: 'Smoking'})
# คำนวณจำนวนแต่ละกลุ่มของความเสี่ยงตาม Smoking Group
risk_counts = df_copy.groupby(['Smoking Group', 'Heart_Attack_Risk']).size().reset_index(name='Count')

# คำนวณเปอร์เซ็นต์ของแต่ละกลุ่มใน Smoking Group
risk_counts['Percentage'] = risk_counts.groupby('Smoking Group')['Count'].transform(lambda x: x / x.sum() * 100)

risk_counts['Heart_Attack_Risk'] = pd.Categorical(risk_counts['Heart_Attack_Risk'], 
                                                  categories=['Low', 'Moderate', 'High'], 
                                                  ordered=True)

# สร้างกราฟแท่ง
plt.figure(figsize=(7, 5))
sns.barplot(data=risk_counts, x='Smoking Group', y='Percentage', hue='Heart_Attack_Risk',hue_order=['Low', 'Moderate', 'High'], palette="RdPu")

plt.title('Heart Attack Risk Percentage by Smoking Status')
plt.xlabel('Smoking')
plt.ylabel('Percentage (%)')
plt.legend(title='Heart Attack Risk')

plt.show()
```

```python
df_copy['Alcohol_Consumption Group'] = df_copy['Alcohol_Consumption'].map({0: 'not Alcohol_Consumption', 1: 'Alcohol_Consumption'})
total_counts = df_copy['Alcohol_Consumption Group'].value_counts()

# นับจำนวนคนที่มีความเสี่ยงระดับกลางและสูงในแต่ละกลุ่ม
risk_counts = df_copy[df_copy['Heart_Attack_Risk'].isin(['High'])]['Alcohol_Consumption Group'].value_counts()

# คำนวณเปอร์เซ็นต์ของคนที่มีความเสี่ยงระดับกลางและสูง
risk_percent = (risk_counts / total_counts) * 100

# แสดงผลลัพธ์
print("อัตราความเสี่ยงระดับ High:")
print(risk_percent)

# Create the countplot with custom colors
plt.figure(figsize=(7, 5))
sns.countplot(data=df_copy, x='Alcohol_Consumption Group', hue='Heart_Attack_Risk', palette=color_t)
plt.title('Heart_Attack_Risk by Alcohol_Consumption ')
plt.xlabel('Alcohol_Consumption')
plt.ylabel('Count')
plt.legend(title='Heart_Attack_Risk')
plt.show()
```

```python
df_copy['Alcohol_Consumption Group'] = df_copy['Alcohol_Consumption'].map({0: 'not Alcohol_Consumption', 1: 'Alcohol_Consumption'})
# คำนวณจำนวนแต่ละกลุ่มของความเสี่ยงตาม Smoking Group
risk_counts = df_copy.groupby(['Alcohol_Consumption Group', 'Heart_Attack_Risk']).size().reset_index(name='Count')

# คำนวณเปอร์เซ็นต์ของแต่ละกลุ่มใน Smoking Group
risk_counts['Percentage'] = risk_counts.groupby('Alcohol_Consumption Group')['Count'].transform(lambda x: x / x.sum() * 100)

# เรียงลำดับ Heart_Attack_Risk
risk_counts['Heart_Attack_Risk'] = pd.Categorical(risk_counts['Heart_Attack_Risk'], 
                                                  categories=['Low', 'Moderate', 'High'], 
                                                  ordered=True)

# สร้างกราฟแท่ง
plt.figure(figsize=(7, 5))
sns.barplot(data=risk_counts, x='Alcohol_Consumption Group', y='Percentage', hue='Heart_Attack_Risk',hue_order=['Low', 'Moderate', 'High'], palette="RdPu")

plt.title('Heart Attack Risk Percentage by Alcohol_Consumption Status')
plt.xlabel('Alcohol_Consumption')
plt.ylabel('Percentage (%)')
plt.legend(title='Heart Attack Risk')

plt.show()
```
