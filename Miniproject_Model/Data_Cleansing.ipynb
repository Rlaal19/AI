```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("./heart_attack_risk_dataset.csv")

data.info()
```

```python
y = data['Heart_Attack_Risk']
df_y = pd.DataFrame(y)
data = data.drop('Heart_Attack_Risk', axis=1)
df_y['Heart_Attack_Risk'] = df_y.Heart_Attack_Risk.map({'Low':0,'Moderate':1,'High':2})
df_y.head(10)
```

```python
standard = ['Age','BMI','Cholesterol_Level','Resting_BP','Heart_Rate','Max_Heart_Rate_Achieved']
data_n = data[standard]
data_D = (data_n-data_n.mean())/data_n.std()
data = data.drop(standard, axis=1)
data.info()
```

```python
# Convert the data into a pandas DataFrame
df = pd.DataFrame(data)
col = ['Gender','Smoking', 'Alcohol_Consumption','Physical_Activity_Level','Diabetes','Hypertension','Family_History','Stress_Level','Chest_Pain_Type','Thalassemia','Fasting_Blood_Sugar','ECG_Results','Exercise_Induced_Angina']
dummies = pd.get_dummies(df,columns=col)
df_onehot = pd.concat([df, dummies], axis = 1)
df_onehot = df_onehot.drop(columns=col)
df_onehot[standard] = data_D
df_onehot['Heart_Attack_Risk'] = df_y
nan_count = df_onehot.isnull().sum().sum()
print('Number of NaN values:', nan_count)
df_onehot.to_csv('cleandata_one_hot_only.csv') 
df_onehot.info()
```
