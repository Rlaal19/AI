```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder
data = pd.read_csv("./heart_disease.csv")
data = data.dropna()
nan_count = data.isnull().sum().sum()
print('Number of NaN values:', nan_count)
data.head(10)
```

```python
target = 'Heart_stroke'
y = data[target]
df_y = pd.DataFrame(y)
data = data.drop(target, axis=1)
df_y[target] = df_y.Heart_stroke.map({'No':0,'yes':1})
nan_count = data.isnull().sum().sum()
print('Number of NaN values:', nan_count)
df_y.head(10)
```

```python
standard = ['age', 'cigsPerDay','totChol','sysBP','diaBP','BMI','heartRate','glucose']
data_n = data[standard]
data_D = (data_n-data_n.mean())/data_n.std()
data = data.drop(standard, axis=1)
nan_count = data.isnull().sum().sum()
print('Number of NaN values:', nan_count)
data_D.head()
```

```python
data.head()
df = pd.DataFrame(data)
col = ['Gender','education','currentSmoker','BPMeds','prevalentStroke','prevalentHyp','diabetes']
dummies = pd.get_dummies(df,columns=col)
df_onehot = pd.concat([df, dummies], axis = 1)
df_onehot = df_onehot.drop(columns=col)
df_onehot[standard] = data_D
df_onehot[target] = df_y
nan_count = df_onehot.isnull().sum().sum()
print('Number of NaN values:', nan_count)
df_onehot.to_csv('cleandata_one_hot_only.csv') 
```
